```python
from openai import OpenAI
import json

client = OpenAI()

# Tool schema the model can use
tools = [
    {
        "type": "function",
        "name": "get_horoscope",
        "description": "Get today's horoscope for an astrological sign.",
        "parameters": {
            "type": "object",
            "properties": {
                "sign": {
                    "type": "string",
                    "description": "An astrological sign like Taurus or Aquarius"
                }
            },
            "required": ["sign"]
        }
    }
]

# Real Python function
def get_horoscope(sign):
    return f"{sign}: Next Tuesday you will befriend a baby otter."

# Conversation history
input_items_1 = [
    {
        "role": "user",
        "content": "What is my horoscope? I am an Aquarius."
    }
]

# Step 1: Ask the model
response = client.responses.create(
    model="gpt-5",
    tools=tools,
    input=input_items_1
)

# Add model output to conversation history
input_items_1 += response.output

# Step 2: Execute any function calls returned by the model
for item in response.output:
    if item.type == "function_call" and item.name == "get_horoscope":
        args = json.loads(item.arguments)   # item.arguments is JSON text
        result = get_horoscope(args["sign"])

        # Send tool result back to the model
        input_items_1.append({
            "type": "function_call_output",
            "call_id": item.call_id,
            "output": json.dumps({
                "horoscope": result
            })
        })

# Step 3: Ask the model again so it can produce the final user-facing answer
final_response = client.responses.create(
    model="gpt-5",
    instructions="Respond only with a horoscope generated by a tool.",
    input=input_items_1
)

print(final_response.output_text)
```

```python
#Working without agent with tools using dictionary propmpt

from langchain_openai import ChatOpenAI
from langchain.tools import tool, ToolRuntime
from langchain_core.messages import ToolMessage

@tool("calculator")
def calc(number1: int, number2: int, operation_type: str) -> str:
    '''Performs arithmetic calculations. 
       If user gives 1 + 2, pass 1 to param1, 2 to param2, and + to operation_type
    '''
    if operation_type == '+':
        return str(number1 + number2)
    elif operation_type == '-':
        return str(number1 - number2)
    elif operation_type == '*':
        return str(number1 * number2)
    elif operation_type == '/':
        return str(number1 / number2)
    else:
        return "Operation not supported"

@tool("weather")
def weather(city: str) -> str:
    '''gives weather information. 
       If user gives city name, pass it to param1
    '''
    return f"The weather in {city} is sunny"

def is_tool(llm_response):
    if(len(llm_response.tool_calls) > 0):
        return True
    return False

def execute_tool(llm_response,messages):
    messages.append({
        "role": "assistant",
        "content": llm_response.content,
        "tool_calls": llm_response.tool_calls 
    })
    tool_info = llm_response.tool_calls[0]
    tool_name = tool_info["name"]
    tool_id=tool_info["id"]
    tool_args=tool_info["args"]
    message_obj = {
        "role":"tool",
        "tool_call_id":tool_id,
        "name":tool_name
    }

    if tool_name == "calculator":
        tool_res = calc.invoke(tool_args)
        message_obj["content"] = tool_res
        messages.append(message_obj)   
    if tool_name == "weather":
        tool_res = weather.invoke(tool_args)
        message_obj["content"] = tool_res
        messages.append(message_obj)   
        

messages = []
model = ChatOpenAI(model="gpt-4o")
model_with_tools = model.bind_tools([calc,weather])

messages.append({"role":"user","content":"Hi this is Feroz practicing langchain"})
result= model_with_tools.invoke(messages)

if is_tool(result) == True:
    execute_tool(result,messages)
else:
    messages.append({"role":"assistant","content":result.content})

messages.append({"role":"user","content":"can you tell my name?"})
result= model_with_tools.invoke(messages)

if is_tool(result) == True:
    execute_tool(result,messages)
else:
    messages.append({"role":"assistant","content":result.content})

messages.append({"role":"user","content":"How much is 10+20?"})
result= model_with_tools.invoke(messages)

if is_tool(result) == True:
    execute_tool(result,messages)
    result = model_with_tools.invoke(messages)
    messages.append({"role":"assistant","content":result.content})
else:
    messages.append({"role":"assistant","content":result.content})

messages.append({"role":"user","content":"How is weather in Singapore?"})
result= model_with_tools.invoke(messages)

#print(messages)
if is_tool(result) == True:
    execute_tool(result,messages)
    result = model_with_tools.invoke(messages)
    messages.append({"role":"assistant","content":result.content})
else:
    messages.append({"role":"assistant","content":result.content})


for message in messages:
    #print(message)
    role = message.get("role","")
    content = message.get("content","")
    tool_calls = message.get("tool_calls","")
    tool_call_name = ""
    if len(tool_calls) > 0:
      tool_call_name = tool_calls[0]["name"]
      print(f"[AI SUGGESTED TOOL]: {tool_call_name}")   
    else:
      if role!="" and content!="":
        print(f"[{role.upper()}]: {content}")
```
